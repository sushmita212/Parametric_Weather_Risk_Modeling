```python
import re

# 1. ID columns
ID_cols = [col for col in df_year.columns if re.search(r'_ID$', col.upper())]

# 2. Timing columns
timing_cols = [col for col in df_year.columns 
               if any(key in col.upper() for key in ['YEAR', 'DATE', 'TIME', 'MONTH', 'DAY'])]

# 3. Location columns
location_cols = [col for col in df_year.columns 
                 if any(key in col.upper() for key in ['STATE', 'LAT', 'LON', 'LOCATION', 'RANGE', 'AZIMUTH', 'CZ_'])]

# 4. Damage columns
damage_cols = [col for col in df_year.columns 
               if any(key in col.upper() for key in ['DAMAGE', 'FATAL', 'DEATH', 'INJUR', 'LOSS', 'CROP'])]

# 5. Severity columns
severity_cols = [col for col in df_year.columns 
                 if any(key in col.upper() for key in ['MAGNITUDE', 'SCALE', 'WIDTH', 'LENGTH', 'CAUSE'])]

# Just to check overlaps (sometimes a col fits 2 groups)
all_grouped = set(ID_cols + timing_cols + location_cols + damage_cols + severity_cols)
ungrouped = [col for col in df_year.columns if col not in all_grouped]

print("ID:", ID_cols)
print("Timing:", timing_cols)
print("Location:", location_cols)
print("Damage:", damage_cols)
print("Severity:", severity_cols)
print("Ungrouped (check manually):", ungrouped)
```
